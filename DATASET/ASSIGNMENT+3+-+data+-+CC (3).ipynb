### IMPORTING LIBRARIES 

```python
import  urllib
import bs4
from bs4 import BeautifulSoup as bs
import os
import argparse
from urllib.request import Request, urlopen
import pandas as pd
from pandas import DataFrame as df
import json
```

### PARSING THE DATASET 

```python
url = ('https://seekingalpha.com/article/4212049-netflix-inc-nflx-ceo-reed-hastings-q3-2018-results-earnings-call-transcript?part=single')

netflix_transcript= Request(url,headers={'User-Agent': 'Mozilla/5.0'})
transcript = urlopen(netflix_transcript).read()

netflix_transcript_2 = bs(transcript,'html.parser')
```

```python
script_1 = netflix_transcript_2.findAll('p',attrs={'p p1'})
script_2 = netflix_transcript_2.findAll('p',attrs={'p p2'})
script_3 = netflix_transcript_2.findAll('p',attrs={'p p3'})
script_4 = netflix_transcript_2.findAll('p',attrs={'p p4'})
script_5 = netflix_transcript_2.findAll('p',attrs={'p p5'})
script_6 = netflix_transcript_2.findAll('p',attrs={'p p6'})
script_7 = netflix_transcript_2.findAll('p',attrs={'p p7'})
script_8 = netflix_transcript_2.findAll('p',attrs={'p p8'})
script_9 = netflix_transcript_2.findAll('p',attrs={'p p9'})
script_10 = netflix_transcript_2.findAll('p',attrs={'p p10'})
```

```python
scripts = (script_1,script_2,script_3,script_4,script_5,script_6,script_7,script_8,script_9,script_10)
```

### SPLITTING THE DATASET INTO PARAGRAPHS 

```python
split_para=[]
for i in range(0,10):
    split=[]
    for j in range(len(scripts[i])):
        txt = scripts[i][j].text.strip()
        if len(txt.split(' ')) > 8:
            split.append(txt)
    split_para.append(split)
```

```python
split_para
```

### CONVERTING IT INTO DATAFRAME

```python
dataframe = []
for i in range(len(split_para)):
    splits = pd.DataFrame(split_para[i],columns=['text'])
    dataframe.append(splits)
```

```python
dataframes = (dataframe[0],dataframe[1],dataframe[2],dataframe[3],dataframe[4],dataframe[5],dataframe[6],dataframe[7],dataframe[8],dataframe[9])
```

```python
dataframes = pd.concat(dataframes,ignore_index=True)
```

```python
dataframes
```

```python
dataframes.index = dataframes.index.map(str)
dict = dataframes.to_dict()
json = json.dumps(dict,indent=2)
f = open('FANG.json',"w")
f.write(json)
f.close()
```
